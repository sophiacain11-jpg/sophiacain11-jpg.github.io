Import Data Frames

```python
import pandas as pd

# load in csv
df_laborforce = pd.read_csv("LaborForceParticipation.csv")

# create additional column with labor force participation rate as a percentage
df_laborforce["Overall Percentage"] = df_laborforce["Overall"] * 100

# print out the first few rows of data frame
df_laborforce.head()
```

```python
# load in csv
df_annualinflation = pd.read_csv("AnnualInflation.csv")

# create additional column with inflation rate as a percentage
df_annualinflation["Overall Percentage"] = df_annualinflation["Overall"] * 100

# print out the first few rows of data frame
df_annualinflation.head()
```

Create Visualization

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(
        x=df_laborforce["date"],
        y=df_laborforce["Overall Percentage"],
        name="Labor Force Participation %",
    ),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(
        x=df_annualinflation["date"],
        y=df_annualinflation["Overall Percentage"],
        name="Annual Inflation %",
    ),
    secondary_y=True,
)

# Add figure title
fig.update_layout(
    title_text="Labor Force Participation vs. Annual Inflation, United States"
)

# Set x-axis title
fig.update_xaxes(title_text="Year")

# Set y-axes titles
fig.update_yaxes(title_text="Labor Force Participation %", secondary_y=False)
fig.update_yaxes(title_text="Inflation Rate %", secondary_y=True)

fig.show()
```
