```python
import pandas as pd

df = pd.read_csv("WDICSV.csv")

df_electricity = df[df["Indicator Name"] == "Access to electricity (% of population)"]

df_electricity.head()
```

```python
electricity_by_country_by_year = (
    df_electricity.drop(columns=["Indicator Name", "Indicator Code"])
    .melt(
        id_vars=["Country Name", "Country Code"],
        var_name="Year",
        value_name="Access to electricity (% of population)",
    )
    .dropna()
)

electricity_by_country_by_year
```

```python
import plotly.express as px

fig = px.line(
    electricity_by_country_by_year,
    x="Year",
    y="Access to electricity (% of population)",
    color="Country Name",
    title="Access to electricity over time",
)
fig.show()
```

```python
happiness = pd.read_excel(
    "https://files.worldhappiness.report/WHR25_Data_Figure_2.1v3.xlsx"
)
happiness
```

```python
fig = px.line(
    happiness,
    x="Year",
    y="Life evaluation (3-year average)",
    color="Country name",
    title="World Happiness",
)
fig = fig.show()
```

```python
country_name = "Albania"

albania_electricity = electricity_by_country_by_year[
    electricity_by_country_by_year["Country Name"] == country_name
]
albania_electricity
```

```python
albania_happiness = happiness[happiness["Country name"] == country_name]
albania_happiness
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots

# Create figure with secondary y-axis
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces
fig.add_trace(
    go.Scatter(
        x=albania_electricity["Year"],
        y=albania_electricity["Access to electricity (% of population)"],
        name="Access to electricity",
    ),
    secondary_y=False,
)

fig.add_trace(
    go.Scatter(
        x=albania_happiness["Year"],
        y=albania_happiness["Life evaluation (3-year average)"],
        name="Life evaluation (3-year average)",
    ),
    secondary_y=True,
)

# Add figure title
fig.update_layout(title_text="Access to electricity vs. Happiness, Albania")

# Set x-axis title
fig.update_xaxes(title_text="Year")

# Set y-axes titles
fig.update_yaxes(title_text="% of population", secondary_y=False)
fig.update_yaxes(title_text="Happiness score", secondary_y=True)

fig.show()
```
